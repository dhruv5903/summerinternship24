```python
def extract_text_from_pdf(pdf_path):
    text = ""
    with pymupdf.open(pdf_path) as doc:
        for page_num in range(len(doc)):
            page = doc.load_page(page_num)
            text += page.get_text()
    return text

def extract_text_from_txt(txt_path):
    with open(txt_path, "r") as f:
        return f.read()

def extract_text_from_files(file_path):
    texts = []
    if file_path.endswith(".pdf"):
        texts.append(extract_text_from_pdf(file_path))
    elif file_path.endswith(".txt"):
        texts.append(extract_text_from_txt(file_path))
    return texts

# Directory containing PDF and text files
file_path = "/content/drive/MyDrive/COVIDTEXTCORPUS.txt"
texts = extract_text_from_files(file_path)
```

```python
def generate_embeddings(text1):
  inputs = tokenizer(text1, return_tensors="pt", padding=True, truncation=True)
  outputs = model(**inputs)
  embedding = outputs.last_hidden_state.mean(dim=1).detach().numpy().flatten()
  return embedding
```

```python
doc_web_embeddings=[]
for docji in clean_web_chunk:
  doc_web_embeddings.append(generate_embeddings(docji))
doc_web_embeddings
```

```python
import numpy as np
for doc, embedding in zip(docs_clean_web_chunk, doc_web_embeddings):
    key = f"doc:{doc['id']}"
    redis_client.hset(key, mapping={
        "content": doc["content"],
        "embedding": np.array(embedding).tobytes()  # Convert embedding to bytes
    })
```

```python
def retrieve(query):
  query_embedding = generate_embeddings([query])
  keys = redis_client.keys("doc:*")
  results = []

  for key in keys:
      content = redis_client.hget(key, "content")
      embedding = np.frombuffer(redis_client.hget(key, "embedding"), dtype=np.float32)
      similarity = np.dot(query_embedding, embedding) / (np.linalg.norm(query_embedding) * np.linalg.norm(embedding))
      results.append((content, similarity))

  # Sort results by similarity
  results = sorted(results, key=lambda x: x[1], reverse=True)[:3]
  return [result[0] for result in results]
```
